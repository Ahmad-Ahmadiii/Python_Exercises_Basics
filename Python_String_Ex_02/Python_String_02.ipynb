 1) Write a program that prompts for a file name, then opens that file and reads through the file, and print the contents of the file in upper case. Use the file **words.txt** in folder
 
___
#### <span style='color:red'>Hint :</span>

* **open():** open(file name, mode='r') -- make a handle to that specific file
* here file name.txt = words.txt
* mode = 'r' means open a file for reading.
* mode = 'w' means open a file for writing.
* mode = 'a' means open a file for appending.

#### <span style='color:green'> Note: </span>



___

```python
# reading a txt file named (words) and printing all of its lines.
# pormpt for file name:
file_name = input('Ener file name: ')
file_name = file_name + '.txt'

print() # creating gap to have a clean output.
print('The original text is:', '\n')

# using open to create a hanle to file:
file_handle = open(file_name, mode = 'r')
# using for loop to iterate through each line in the words.txt file.
for line in file_handle:
    print(line)
```

```python
# reading a txt file named (words.txt) and uppercase all of its contents.
# prompt for file name:
file_name = input('Enter file name:')
file_name = file_name + '.txt'
# catching wrong file name by try & except block:
try:
    file_handle = open(file_name)
except:
    print('invalid file name')

print() # creating gap to have a clean output.
print(' The Uppercase of contents is:', '\n')

# using for loop to iterate through lines:
for line in file_handle:
    # using upper() method to uppercase each charachter of current line:
    cap_line = line.upper()
    print(cap_line)
```
