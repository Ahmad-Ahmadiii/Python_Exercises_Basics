```python
# opening (mbox_short.txt) file, and reading through each line:
file_name = input('Enter file name: ')
file_name = file_name + '.txt'
# making file handle object, using open():
file_handle = open(file_name, 'r')
hour_dict = {}
hour_list = []
# reading through each line of (mbox_short.txt):
for line in file_handle:
    # finding line which stars whith : 'From ':
    if line.startswith('From '):
        # finding hour using split() method:
        email_hour_complete = line.split()
        email_hour_complete = email_hour_complete[5]
        # finding partial part of hour:
        email_hour = email_hour_complete.split(':')
        email_hour = email_hour[0]
        hour_list.append(email_hour)
hour_list.sort()
for hour in hour_list:
    hour_dict[hour] = hour_dict.get(hour, 0) + 1
# printing results:      
print('__________')
print ('email hours dictionary:\n', hour_dict)
print('__________')
for hour, value in hour_dict.items():
    print(hour, value)
        
```
